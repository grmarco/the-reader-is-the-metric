```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import data_loader
import config
```

```python
from config import selected_features
```

```python
df = data_loader.load_text_metrics_data(config.stories_metrics_filepath)
# Upload data
df_inliers, features_inliers = data_loader.calculate_features(df, 
                                                              feature_cols=selected_features, 
                                                              normalization='standard', 
                                                              remove_outliers_flag=False)
```

```python
# Assume df_inliers and features_cols are already defined
# Calculate the correlation
correlation = df_inliers[selected_features].corr()

# the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='RdYlGn', fmt=".2f")
plt.title('Heatmap of text features correlation')

plt.savefig('figures/heatmap_correlation.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
